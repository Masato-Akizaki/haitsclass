```python
# 線形回帰モデルのクラスを読み込み
from sklearn.linear_model import LinearRegression

# 線形回帰のインスタンスを生成
lr = LinearRegression()
```

```python
X1 = df1.loc[:, ['OverallQual']].values
y1 = df1.loc[:, ['SalePrice']].values

# 単回帰モデルを学習
lr.fit(X1, y1)
```

```python
lr.intercept_
```

```python
lr.coef_
```

```python
# 散布図に近似直線を重ねてみる
plt.scatter(X1, y1)                       # 散布図を表示
plt.plot(X1, lr.predict(X1), color='red') # 回帰直線を表示
plt.title('SalePrice vs GrLivArea')     # グラフのタイトルを表示
plt.xlabel('GrLivArea')               # x軸のラベルを表示
plt.ylabel('SalePrice')                  # y軸のラベルを表示
plt.show()                              # 上記の内容でグラフを表示
```

```python
# 2次関数の変数変換
from sklearn.preprocessing import PolynomialFeatures
quad = PolynomialFeatures(degree=2)    # 2次の多項式規定を生成
X2 = quad.fit_transform(X1) # 生成した基底関数で変数変換を実行
```

```python
# 線形回帰のインスタンスを生成
model_quad = LinearRegression()
# 変換後の変数を利用して実行
model_quad.fit(X2, y1)
```

```python
# 散布図に近似直線､近似曲線を重ねてみる
plt.scatter(X1, y1, color='lightgray', label='data')               # 散布図を表示
plt.plot(X1, lr.predict(X1), color='red', label='linear')       # 回帰直線を表示
x_quad = quad.fit_transform(X1)                                       # プロットのためのxも2変数変換する
plt.plot(X1, model_quad.predict(X2), color='green', label='quad') # 回帰曲線を表示（2次関数）
plt.xlabel('GrLivArea')                                                  # x軸のラベルを表示
plt.ylabel('SalePrice')                                                   # y軸のラベルを表示
plt.legend(loc = 'upper right')                                      # 凡例を表示
plt.show()                 
```
