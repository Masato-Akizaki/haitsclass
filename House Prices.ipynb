```python
#説明変数'OverallQual'と'GrLivArea'で多項式回帰を行う
```

```python
# 特徴量を抽出
Ov = df1.loc[:, ['OverallQual']].values
Gr = df1.loc[:, ['GrLivArea']].values
y = df1.loc[:, ['SalePrice']].values
```

```python
#説明変数'OverallQual'の2次関数のモデルと3次関数のモデル
```

```python
# 2次関数の変数変換
from sklearn.preprocessing import PolynomialFeatures
quad = PolynomialFeatures(degree=2)    # 2次の多項式規定を生成
Ov_quad = quad.fit_transform(Ov) # 生成した基底関数で変数変換を実行
Ov_quad
```

```python
# ホールド・アウト法によるデータの分割
# random_stateを0に固定してあるので､毎回同じサンプルに分割
Ov_quad_train, Ov_quad_test, _, _ = train_test_split(Ov_quad, y, test_size = 0.3, random_state = 0)
```

```python
# 線形回帰のインスタンスを生成
model_quad = LinearRegression()

# 変換後の変数を利用して実行
model_quad.fit(Ov_quad_train, y_train)
```

```python
# 3次関数の変数変換
cubic = PolynomialFeatures(degree=3)     # 3次の多項式規定を生成
Ov_cubic = cubic.fit_transform(Ov) # 生成した基底関数で変数変換を実行
Ov_cubic
```

```python
# ホールド・アウト法による分割
Ov_cubic_train, Ov_cubic_test, _, _ = train_test_split(Ov_cubic, y, test_size = 0.3, random_state = 0)
```

```python
# 線形回帰のインスタンスを生成
model_cubic = LinearRegression()

# 変換後の変数を利用して実行
model_cubic.fit(Ov_cubic_train, y_train)
```

```python
# 自由度調整済み決定係数
# (決定係数, trainまたはtestのサンプル数, 利用した特徴量の数)
def adjusted(score, n_sample, n_features):
    adjusted_score = 1 - (1 - score) * ((n_sample - 1) / (n_sample - n_features - 1))
    return adjusted_score
```

```python
# 自由度調整済み決定係数をtrainとtestに分けて出力
print('adjusted R^2')
print('')

# 2次関数
print('OverallQual 2次関数')
print('train: %.3f' % adjusted(model_quad.score(Ov_quad_train, y_train), len(y_train), 2))
print('test : %.3f' % adjusted(model_quad.score(Ov_quad_test, y_test), len(y_test), 2))
print('')

# 3次関数
print('OverallQual 3次関数')
print('train: %.3f' % adjusted(model_cubic.score(Ov_cubic_train, y_train), len(y_train), 3))
print('test : %.3f' % adjusted(model_cubic.score(Ov_cubic_test, y_test), len(y_test), 3))
```

```python
#↑テストデータの方が決定係数が高くなった。
```
