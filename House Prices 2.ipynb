```python
#抽出と変数変換
from sklearn.preprocessing import PolynomialFeatures
quad = PolynomialFeatures(degree=2)    # 2次の多項式規定を生成
cubic = PolynomialFeatures(degree=3)     # 3次の多項式規定を生成
fourth = PolynomialFeatures(degree=4)     # 4次の多項式規定を生成

#'OverallQual'
Ov = df1_log.loc[:, ['OverallQual']].values
Ov_quad = quad.fit_transform(Ov)
Ov_cubic = cubic.fit_transform(Ov) 
Ov_fourth = fourth.fit_transform(Ov)

#'GrLivArea'
Gr = df1_log.loc[:, ['GrLivArea']].values
Gr_quad = quad.fit_transform(Gr)
Gr_cubic = cubic.fit_transform(Gr) 
Gr_fourth = fourth.fit_transform(Gr)

#'GarageArea'
Ga = df1_log.loc[:, ['GarageArea']].values
Ga_quad = quad.fit_transform(Ga)
Ga_cubic = cubic.fit_transform(Ga) 
Ga_fourth = fourth.fit_transform(Ga)

#'TotalBsmtSF'
To = df1_log.loc[:, ['TotalBsmtSF']].values
To_quad = quad.fit_transform(To)
To_cubic = cubic.fit_transform(To) 
To_fourth = fourth.fit_transform(To)

#'YearBuilt'
Ye = df1_log.loc[:, ['YearBuilt']].values
Ye_quad = quad.fit_transform(Ye)
Ye_cubic = cubic.fit_transform(Ye) 
Ye_fourth = fourth.fit_transform(Ye)

y = df1_log.loc[:, ['SalePrice_log']].values
```

```python
# 自由度調整済み決定係数
# (決定係数, trainまたはtestのサンプル数, 利用した特徴量の数)
def adjusted(score, n_sample, n_features):
    adjusted_score = 1 - (1 - score) * ((n_sample - 1) / (n_sample - n_features - 1))
    return adjusted_score
```
