```python
import matplotlib.pyplot as plt
% matplotlib inline
```

```python
# サンプルとなるデータセットの作成
import numpy as np
np.random.seed(0)                       # 乱数生成器の状態を指定
x_1 = np.random.normal(50, 10, 100)     # 平均50､標準偏差10の正規分布に従うデータ点を100個生成
y_1 = x_1 + np.random.normal(0, 5, 100) # xに対応するyの値を算出
```

```python
x_1[:24]
```

```python
y_1[:24]
```

```python
plt.scatter(x_1,y_1)
plt.show()
```

```python
plt.scatter(x_1,y_1)
plt.title('The relationship between x and y')
plt.xlabel('x')
plt.ylabel('y')
plt.show()
```

```python
plt.scatter(x_1,y_1,color='y')
plt.title('The relationship between x and y')
plt.xlabel('x')
plt.ylabel('y')
plt.show()
```

```python
# 新しい系列のデータを作成
# 中のコードは気になくても大丈夫です
x_2 = np.random.normal(30, 5, 100)                  # 平均30､標準偏差5の正規分布に従うデータ点を100個生成
y_2 = 2.5 * x_2 + np.random.normal(0, 10, len(x_2)) # xに対応するyの値を算出
```

```python
plt.scatter(x_1,y_1,color='r')
plt.scatter(x_2,y_2,color='b')
plt.title('The relationship between x and y')
plt.xlabel('x')
plt.ylabel('y')
plt.show()
```

```python
plt.scatter(x_1,y_1,color='r',marker='s')
plt.scatter(x_2,y_2,color='b',marker='*')
plt.title('The relationship between x and y')
plt.xlabel('x')
plt.ylabel('y')
plt.show()
```

```python
plt.scatter(x_1,y_1,color='r',marker='s',label='data1')
plt.scatter(x_2,y_2,color='b',marker='x',label='data2')
plt.title('The relationship between x and y')
plt.xlabel('x')
plt.ylabel('y')
plt.legend(loc='lower right') 
plt.show()
```

```python
plt.scatter(x_1,y_1,color='r',marker='s',label='data1')
plt.scatter(x_2,y_2,color='b',marker='x',label='data2')
plt.title('The relationship between x and y')
plt.xlabel('x')
plt.ylabel('y')
plt.legend(loc='lower right') 
plt.xticks([20,50,80])
plt.yticks([20,50,80,110])
plt.show()
```

```python
plt.scatter(x_1,y_1,color='r',marker='s',label='data1')
plt.scatter(x_2,y_2,color='b',marker='x',label='data2')
plt.title('The relationship between x and y')
plt.xlabel('x')
plt.ylabel('y')
plt.legend(loc='lower right') 
plt.xticks(np.arange(10,80,5))
plt.yticks(np.arange(10,130,10))
plt.show()
```

```python
plt.scatter(x_1,y_1,color='r',marker='s',label='data1')
plt.scatter(x_2,y_2,color='b',marker='x',label='data2')
plt.title('The relationship between x and y')
plt.xlabel('x')
plt.ylabel('y')
plt.legend(loc='lower right') 
plt.xticks(np.arange(10,80,5))
plt.yticks(np.arange(10,130,10))
plt.grid()
plt.show()
```

```python
x_3=np.arange(-10,10,0.1)
y_3=0.001*(x_3**3+x_3**2+x_3+1)
plt.plot(x_3,y_3)
plt.show()
```

```python
plt.plot(x_3,y_3,label='cubic')
plt.title('a cubic function')
plt.xlabel('x')
plt.ylabel('y')
plt.legend(loc='upper left') 
plt.xticks(np.arange(-10, 11, 1))
plt.yticks(np.arange(-1, 1.5, 0.25))
plt.grid()
plt.show()    
```

```python
plt.plot(x_3,y_3,label='cubic')
plt.title('a cubic function')
plt.xlabel('x')
plt.ylabel('y')
plt.legend(loc='upper left') 
plt.xticks(np.arange(-10, 11, 1))
plt.yticks(np.arange(-1, 1.5, 0.25))
plt.grid()
plt.hlines([0],-10,10,linestyle='dashed',color='gray')
plt.vlines([0],-1,1,linestyle='dashed',color='gray')
plt.show()    
```
